```python
import subprocess, sys

packages = [
    "torch",
    "numpy",
    "gymnasium",
    "matplotlib",
    "pandas",
    "scipy",
    "tqdm",
]

for pkg in packages:
    print(f"Installing {pkg}...")
    subprocess.run([sys.executable, "-m", "pip", "install", pkg, "-q"])
    print(f"  ✅ {pkg} done")

print("\n🚀 All libraries installed successfully!")
```
